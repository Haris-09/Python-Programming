```python
dict1 = {
    "name":['ali', 'sara', 'zara', 'hamza', 'sohaib', 'maria'],
    "Maths marks":[56,90,78,95,56,100],
    "Science marks":[60,89,99,89,56,89]
}
df5=pd.DataFrame(dict1)
print(df5)
```

```python
from sklearn.preprocessing import StandardScaler
ss1 = StandardScaler()
ss2=ss1.fit_transform(df5[['Maths marks', 'Science marks']])
ss3=pd.DataFrame(ss2, columns=['Maths marks', 'Science marks'])
ss3=pd.concat([df5['name'], ss3], axis=1)
print(ss3)
```

```python
from sklearn.preprocessing import MinMaxScaler
ss1 = MinMaxScaler()
ss2=ss1.fit_transform(df5[['Maths marks', 'Science marks']])
ss3=pd.DataFrame(ss2, columns=['Maths marks', 'Science marks'])
ss3=pd.concat([df5['name'], ss3], axis=1)
print(ss3)
```

```python
from sklearn.preprocessing import Normalizer
ss1 = Normalizer()
ss2=ss1.fit_transform(df5[['Maths marks', 'Science marks']])
ss3=pd.DataFrame(ss2, columns=['Maths marks', 'Science marks'])
ss3=pd.concat([df5['name'], ss3], axis=1)
print(ss3)
```
