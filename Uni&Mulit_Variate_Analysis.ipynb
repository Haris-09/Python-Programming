```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Univariate Analysis

```python
data = {
    "Name": ["Areeba", "Ali", "Sana", "Bilal", "Asma", "Junaid", "Saba", "Ahmed"],
    "Age": [22, 28, 26, 30, 24, 35, 27, 29],
    "Gender": ["Female", "Male", "Male", "Male", "Female", "Male", "Female", "Male"],
    "Salary": [50000, 70000, 68000, 75000, 52000, 85000, 63000, 70000]
}

df = pd.DataFrame(data)
```

```python
sns.boxplot(x=df['Salary'])
plt.show()
```

```python
sns.histplot(df['Salary'], kde=True, bins=5)
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.countplot(x='Gender', data=df, palette='deep')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

### Multi-Variate Analysis

```python
sns.scatterplot(data=df, x='Age', y='Salary', palette='deep', hue='Gender')
plt.title('Age vs Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

```python
sns.pairplot(df, hue='Gender')
plt.show()
```
