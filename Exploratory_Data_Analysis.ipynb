```python
salaries = pd.read_csv(file_path+'/salary_data.csv')
salaries.head()
```

```python
sns.boxplot(x='Experience', y='Salary_USD', data=salaries, palette='viridis', hue='Experience')
```

```python
salaries.isna().sum()
```

```python
len(salaries)
```

```python
threshold = 45
column_to_remove = salaries.isna().sum()>=threshold
print(column_to_remove)
```

```python
# use one of the lines below to remove the columns which contains empty values above threshold both logics works
# salaries = salaries.drop(columns=column_to_remove[column_to_remove].index)
salaries = salaries.drop(columns=salaries.columns[column_to_remove])
print(salaries.isna().sum())
```

```python
numerical_cloumns = salaries.select_dtypes("number").columns
print(numerical_cloumns)
```

```python
categorical_cloumns = salaries.select_dtypes(exclude="number").columns
print(categorical_cloumns)
```

```python
# fill the missing values with mode for categorical columns
for column in categorical_cloumns:
    salaries[column] = salaries[column].fillna(salaries[column].mode()[0])
```

```python
# fill the missing values with mean for numerical columns
for column in numerical_cloumns:
    salaries[column] = salaries[column].fillna(salaries[column].mean())
```

```python
print(salaries.isna().sum())
```

```python
# mode doesn't work in groupby function as it can return multiple values so used the lambda function in agg function
salaries.groupby('Employment_Status')[categorical_cloumns].agg(lambda x: x.mode().iloc[0] if not x.mode().empty else None)
```

```python
salary_dict = salaries.groupby('Employment_Status')['Working_Year'].mean().to_dict()
print(salary_dict)
```

```python
salaries['Working_Year'] = salaries['Working_Year'].fillna(salaries['Employment_Status'].map(salary_dict))
```
