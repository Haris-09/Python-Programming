```python
import pandas as pd

file_path = '/content/drive/My Drive/NAVTCC: Data Science & AI' # Replace with your file path
books = pd.read_csv(file_path+'/books_01.csv')
books.head()
```

```python
# How many types of Books we have
books.Genre.unique()
```

```python
books.Genre.nunique()
```

```python
books.describe()
```

```python
# How many books are there in each category of Genre
books.Genre.value_counts()
```

```python
# Average grading of each category in genre
books.groupby('Genre')['Ratings'].mean()
```

```python
books.groupby('Genre').agg(
    grp_mean = ('Ratings', 'mean'),
    grp_std = ('Ratings', 'std'),
    grp_median = ('Ratings', 'median')
    )
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.histplot(data=books, x='Ratings', binwidth=0.1)
plt.show()
```

```python
books['Ratings'] = books['Ratings'].astype(int)
books.dtypes
```
